```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import scipy.io
features = scipy.io.loadmat('DenseNetricofeat.mat')
```

```python
features.items()
```

```python
data=features["interdicenetrico"]
```

```python
data
```

```python
df = pd.DataFrame()
```

```python
dataf_new = pd.concat([df, pd.DataFrame(data)], axis=1)
```

```python
dataf_new
```

```python
from sklearn import preprocessing
dataf_normalized = preprocessing.normalize(dataf_new)
```

```python
dataf_normalized
```

```python
dataf_normalizedframe = pd.concat([df, pd.DataFrame(dataf_normalized)], axis=1)
```

```python
dataf_normalizedframe
```

```python
labels=scipy.io.loadmat('labels.mat')
```

```python
labels.items()
```

```python
Classes=labels["labels"]
Classes
```

```python
df1 = pd.DataFrame()
```

```python
df_new = pd.concat([df1, pd.DataFrame(Classes)], axis=1)
```

```python
df_new
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV

from sklearn import tree
clf = AdaBoostClassifier(learning_rate= 0.01, n_estimators=100)
A=np.mean(cross_val_score(clf,dataf_normalizedframe,df_new,cv=5,scoring="accuracy"))
r=np.mean(cross_val_score(clf,dataf_normalizedframe,df_new,cv=5,scoring="recall_weighted"))
p=np.mean(cross_val_score(clf,dataf_normalizedframe,df_new,cv=5,scoring="precision_weighted"))
print('A=',A*100)
print('r=',r*100)
print('p=',p*100)
```
